```python
from collections import Counter

def Most_Common(lst):
    data = Counter(lst)
    if data.most_common()[0][1] == data.most_common()[-1][1]:
        if data.most_common()[0][0] == "1":
            return data.most_common()[0][0]
        else:
            return data.most_common()[-1][0]
    else:
        return data.most_common()[0][0]

def Least_Common(lst):
    data = Counter(lst)
    if data.most_common()[0][1] == data.most_common()[-1][1]:
        if data.most_common()[0][0] == "0":
            return data.most_common()[0][0]
        else:
            return data.most_common()[-1][0]
    else:
        return data.most_common()[-1][0]
```

```python
gamma = ""
epsilon = ""
for item in list_of_lists:
    gamma = gamma + Most_Common(item)
    epsilon = epsilon + Least_Common(item)

print(gamma, epsilon)
```

```python
# packaged solution to binary conversion
print(int(gamma, 2)*int(epsilon,2))
```

```python
## Created binary to decimal function
def binary_to_decimal(binary):
    char_len = len(binary)
    bit_size = 1
    value = 0
    for char in reversed(binary):
        if char == "1":
            value += bit_size
        bit_size = bit_size *2
    
    return value
```
