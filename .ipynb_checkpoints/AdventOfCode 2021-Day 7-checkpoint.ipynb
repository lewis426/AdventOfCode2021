## Day 7 Part 1

```python
test_df = pd.read_csv(r'\\corp-not-vs12\V12PERSON\HR Analysis Team\01_HRAnalysisTeam\106_Lewis\AdventOfCode2021\Day7\TestInput.csv', header=None)
```

```python
test_list = test_df[0].str.split(',')[0]
test_list = [int(x) for x in test_list]
```

```python
#print(test_list)
```

```python
# Get bounds of number_list to iterate over
high_bound = max(number_list)
low_bound = min(number_list)
# Initiate dictionary to capture fuel costs
fuel_dict = {}
```

```python
# Loop over all possible horizontal values, pull total fuel cost for each value into dict and then pull out cheapest one
for i in range(low_bound, high_bound):
    total_fuel_cost = 0
    for num in test_list:
        fuel_cost = abs(num - i)
        total_fuel_cost += fuel_cost
    fuel_dict[i] = total_fuel_cost
```

```python
# Get the key corresponding to the minimum value from fuel_dict
print(f"The optimal position is {min(fuel_dict, key=fuel_dict.get)} horizontally and the total fuel cost is {fuel_dict[min(fuel_dict, key=fuel_dict.get)]}.")
```

## Day 7 Part 2

```python
test_df = pd.read_csv(r'\\corp-not-vs12\V12PERSON\HR Analysis Team\01_HRAnalysisTeam\106_Lewis\AdventOfCode2021\Day7\TestInput.csv', header=None)
```

```python
test_list = test_df[0].str.split(',')[0]
test_list = [int(x) for x in test_list]
```

```python
#print(test_list)
```

```python
def fuel_cost_per_distance(dist):
    total_fuel_cost = 0
    for i in range(dist):
        total_fuel_cost += i+1
        
    return total_fuel_cost
```

```python
# Get bounds of number_list to iterate over
high_bound = max(number_list)
low_bound = min(number_list)
# Initiate dictionary to capture fuel costs
fuel_dict = {}
```

```python
# Loop over all possible horizontal values, pull total fuel cost for each value into dict and then pull out cheapest one
for i in range(low_bound, high_bound):
    total_fuel_cost = 0
    for num in test_list:
        fuel_cost = fuel_cost_per_distance(abs(num - i))
        total_fuel_cost += fuel_cost
    fuel_dict[i] = total_fuel_cost
```

```python
# Get the key corresponding to the minimum value from fuel_dict
print(f"The optimal position is {min(fuel_dict, key=fuel_dict.get)} horizontally and the total fuel cost is {fuel_dict[min(fuel_dict, key=fuel_dict.get)]}.")
```
